## 1. 加载数据

### 1.1 加载文章的文件

```python
# 全局配置
encoding = 'utf-8'
```

```python
import os

news_path = []
for dirname, _, filenames in os.walk('../data/20231101'):
    for filename in filenames:
        news_path.append(os.path.join(dirname, filename).replace('\\', '/'))

print("新闻总数为: ", len(news_path))
# print("news: ",news_path)
```

### 1.2 尝试读取一下文章内容

```python
from random import sample

for each_path in sample(news_path, k=2):
    with open(each_path, 'r', encoding=encoding) as f:
        news = f.read()
        print(news)
        print('\n\n')
```

### 1.3 使用jieba进行分词

```python
import jieba

from random import sample

for eachpath in sample(news_path, k=1):
    with open(eachpath, 'r', encoding=encoding) as f:
        news = f.read()
        print('原始新闻内容为: ')
        print(news)

        print('以简单模式分割: ')
        print("/".join(jieba.lcut(news)))

        print('以cut_all模式分割: ')
        print("/".join(jieba.lcut(news, cut_all=True)))

        print('以search模式分割: ')
        print("/".join(jieba.lcut_for_search(news)))
```

### 1.4 删除空格、数字和标点符号等内容

```python
# 此处使用zhon包
!pip3 install zhon
```

```python
import re
from zhon.hanzi import punctuation as puncZH
from string import punctuation as puncEN


def removePunctuation(s: str):
    s = re.sub(r"[%s]+" % puncZH, "", s)
    s = re.sub(r"[%s]+" % puncEN, "", s)
    s = re.sub(r"[\d\n\t\r]+", "", s)
    return s
```

```python
from random import choice
news = open(choice(news_path),'r', encoding=encoding).read()
print(news)
```

```python
# 打印去除标点符号的内容
print(removePunctuation(news))
```

```python
# 使用/分隔词
print("/".join(jieba.lcut_for_search(news)))
print("/".join(jieba.lcut_for_search(removePunctuation(news))))
```

```python
# 合并操作
import re
from zhon.hanzi import punctuation as puncZH
from string import punctuation as puncEN


def hasPunctuation(s: str):
    f = False
    f |= bool(re.match(r"[%s]+" % puncZH, s))
    f |= bool(re.match(r"[%s]+" % puncEN, s))
    f |= bool(re.match(r"[\d\n\t\r]+", s))
    return f
```

```python
split_news = jieba.lcut_for_search(news)
removed_split_news = []
for text in split_news:
    if hasPunctuation(text) is False:
        removed_split_news.append(text)

print("/".join(jieba.lcut_for_search(news)))
print("/".join(removed_split_news))
```

```python
# 封装成函数
def getWords(news: str):
    split_news = jieba.lcut_for_search(news)
    removed_split_news = []
    for text in split_news:
        if hasPunctuation(text) is False:
            removed_split_news.append(text)
    return removed_split_news
```

```python
words = []
print(len(news_path))
# for path in sample(newspath, k=40000):
for path in sample(news_path, k=50):
    news = open(path, 'r', encoding=encoding).read()
    words += getWords(news)

print('词语数量: ', len(words))
```

## 2. 生成词云

### 2.1 计算词频

```python
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
print('去重词语的个数：', len(word_count))
```

```python
import pandas as pd

word_ds=pd.DataFrame(word_count,index=[0])
word_ds=word_ds.T
word_ds.columns=['count']
word_ds
```

```python
## 很多没有意义的词
word_ds.sort_values(by=['count'], ascending=False, inplace=True)
word_ds.head(10)
```

## 删除停止语 安装环境
下载https://codeload.github.com/Concyclics/stopwords/zip/refs/heads/master

```python
stopwords = []

for dirname, _, filenames in os.walk('./stopwords-master'):
    for filename in filenames:
        with open(os.path.join(dirname, filename)) as f:
            if filename.endswith('txt'):
                stopwords += list(f.read().split())
    
stopwords.append('年')
stopwords.append('月')
stopwords.append('日')
stopwords.append('上')
stopwords.append('中')
stopwords.append('本报')
stopwords.append('\xa0')

stopwords = list(set(stopwords))
print('统计停止词总数: ', len(stopwords))
```

```python
for stopword in stopwords:
    if stopword in word_count:
        word_count.pop(stopword)

print('不带停止语的唯一单词总数: ', len(word_count))
```

```python
word_ds=pd.DataFrame(word_count,index=[0])
word_ds=word_ds.T
word_ds.columns=['count']

word_ds.sort_values(by=['count'],ascending=False,inplace=True)
word_ds.head(20)
```
